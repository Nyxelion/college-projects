```
# Exercicio  02: A Metamorfose das Grandezas
distancia = float(input("Digite a distância em metros: "))

centimetros = distancia * 100
milimetros = distancia * 1000

print(f"Centímetros: {centimetros} cm | Milímetros: {milimetros} mm")
```

```
#Exercício 03: O Equilíbrio das Finanças Individuais
receita = float(input("Digite sua receita mensal: "))
despesas = float(input("Digite suas despesas fixas: "))

saldo = receita - despesas
positivo = saldo > 0

print(f"Saldo: R$ {saldo}. Terminou positivo? {positivo}")
```

```
# Exercício 04: A Projeção do Tempo Estela
Horas = float(input("digite a quantidade de horas:"))
minutos = Horas * 60
segundos = minutos * 60

print(f"minutos: {minutos:.0f} segundos: {segundos:.0f}")
```

```
#Exercício 05: O Cálculo do Índice de Rendimento Acadêmico
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))

media = (n1*2 + n2*3 + n3*5) / (2+3+5)

print(f"Média Ponderada: {media:.2f}")
```

```
#Exercício 06: A Engenharia de Climatização
Celsius = float(input("Digite a temperatura em Celsius:"))
Fahrenheit = (Celsius * 9/5) + 32
kelvin = Celsius + 273.15
print(f"Fahrenheit:{Fahrenheit}\nKelvin: {kelvin}")
```

```
#Exercício 07: O Simulador de Geometria Analítica
Base = float(input("Digite a base do retangulo "))
Altura = float(input("Digite a Altura do retangulo"))
Área = Base*Altura
Perimetro = (Base*2)+ (Altura*2)
Diagonal = (Base**2 + Altura**2) **0.5
print(f"Área: {Área}| Perimetro: {Perimetro}| Diagonal: {Diagonal}")
```

```
#Exercício 09: A Eficiência Energética de Transmissão
Consumo = float(input("Digite seu consumo:"))
Trabalho = float(input("Digite sua quantidade de Trabalho:"))
perda_percentual = ((Consumo - Trabalho) / Consumo) * 100
Percental = (Trabalho / Consumo) * 100

print(f"Perda de energia: {perda_percentual:.1f}%")

```

```
#Exercício 10: O Árbitro da Lógica Booleana
Vínculo = input("Possui vínculo empregatício? (true/false)").lower() == 'true'
Restrição = input("Possui restrição financeira (True/False)").lower() == 'true'
Renda = input("Renda mensal superior a R$ 3.000,00 (True/False)").lower() == 'true'
credito_concedido = Vínculo and (not Restrição or (Restrição and Renda))

print(f"Crédito Concedido? {credito_concedido}")
```
