```python
#Exercício 4: Calculadora de Desconto
valor = float(input("Digite o valor da compra: R$ "))

if valor > 200:
    desconto = valor * 0.20
    percentual = 20
elif valor >= 100:
    desconto = valor * 0.10
    percentual = 10
else:
    desconto = 0
    percentual = 0

valor_final = valor - desconto

print(f"Desconto aplicado: {percentual}%")
print(f"Valor final: R$ {valor_final:.2f}")
```

```python
#Exercício 5: Categoria de Nadadores

idade = int(input("digite sua idade:"))
if idade < 5:
  print("Idade inválida para competição")
elif idade >= 5 and idade <= 7:
   print("Categoria:Infantil A")
elif idade >= 8 and idade <= 10:
  print("Categoria: Infantil B")
elif idade >= 11 and idade <= 13:
  print("Categoria: Juvenil A")
elif idade >= 14 and idade <= 17:
  print(" Categoria: Juvenil B")
elif idade >= 18:
  print("adulto")
```

```python
#Exercício 6: Sistema de Avaliação Escolar

n1 = float(input("Nota 1:"))
n2 = float(input(" Nota 2:"))
média = (n1 + n2) /2

if média >= 7:
  print("Aprovado")
elif média <= 7 and média >= 5:
  print("Situação:Recuperação")
else:
  print("Reprovado")
```

```python
#Exercício 7: Aprovação de Empréstimo Bancário

valor_imovel = float(input("Digite o valor do imóvel: R$ "))
salario = float(input("Digite o salário: R$ "))
anos = int(input("Em quantos anos vai pagar: "))

meses = anos * 12
prestação = valor_imovel / meses
limite = salario * 0.30

if prestação <= limite:
    print(f"Prestação: R$ {prestação:.2f}. Empréstimo APROVADO! (A prestação não excede 30% do salário, que seria R$ {limite:.2f}).")
else:
    print(f"Prestação: R$ {prestação:.2f}. Empréstimo NEGADO! (A prestação excede 30% do salário, que seria R$ {limite:.2f}).")
```

```python
#Exercício 8: Calculadora de IMC
peso = float(input("Digite o peso (kg): "))
altura = float(input("Digite a altura (m): "))

imc = peso / (altura ** 2)

if imc < 18.5:
    classificacao = "Abaixo do peso"
elif imc < 25:
    classificacao = "Peso normal"
elif imc < 30:
    classificacao = "Sobrepeso"
else:
    classificacao = "Obesidade"

print(f"IMC: {imc:.2f} - Classificação: {classificacao}")
```

```python
#Exercício 9: Logística de Frete

regiao = input("Digite a região (Sul, Sudeste, Centro-Oeste): ").strip().lower()
peso = float(input("Digite o peso do pacote (kg): "))

tabela_frete = {
    "sul": 15,
    "sudeste": 10,
    "centro-oeste": 20
}

if regiao not in tabela_frete:
    print("Região não atendida")
else:
    frete = tabela_frete[regiao]

    if peso > 5:
        frete += 10

    print(f"Valor do frete: R$ {frete:.2f}")
```
